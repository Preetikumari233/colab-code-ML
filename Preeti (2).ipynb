```python
from collections import Counter

def count_char_frequency(input_string):
    """Counts the frequency of each character in a string if its length is > 50."""
    if len(input_string) <= 50:
        print(f"String length ({len(input_string)}) is not more than 50. Skipping character frequency count.")
        return {}

    return Counter(input_string)
```

```python

long_string = "This is a very long string that definitely has more than fifty characters in it, so it should be processed."

print(f"\nString: '{long_string}'")
long_string_frequency = count_char_frequency(long_string)
if long_string_frequency:
    print("Character frequency (long string):")
    for char, count in long_string_frequency.items():
        print(f"'{char}': {count}")


short_string = "This is a short string."

print(f"\nString: '{short_string}'")
short_string_frequency = count_char_frequency(short_string)
if short_string_frequency:
    print("Character frequency (short string):")
    for char, count in short_string_frequency.items():
        print(f"'{char}': {count}")
```

```python
def remove_duplicates_without_set(input_list):
    unique_list = []
    for item in input_list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list


my_list = [1, 2, 2, 3, 4, 4, 5, 1]
cleaned_list = remove_duplicates_without_set(my_list)
print(f"Original list: {my_list}")
print(f"List with duplicates removed: {cleaned_list}")
```
